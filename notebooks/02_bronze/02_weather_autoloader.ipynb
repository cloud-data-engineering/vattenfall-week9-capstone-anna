```python
# Day 1 Bronze ingestion - weather observations

catalog = "vattenfall_dev"
schema = "raw"

landing_path = f"/Volumes/{catalog}/{schema}/landing/weather"
checkpoint_path = f"/Volumes/{catalog}/{schema}/checkpoints/weather_checkpoint"
target_table = f"{catalog}.{schema}.bronze_weather"

print("Landing path:", landing_path)
print("Checkpoint path:", checkpoint_path)
print("Target table:", target_table)
```

```python
# Inspect weather landing files

display(dbutils.fs.ls(landing_path))
```

```python
# Load weather CSV files

weather_df = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(landing_path)
)

display(weather_df)
weather_df.printSchema()
```

```python
# Write weather data to Bronze Delta table

(
    weather_df
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(target_table)
)

print(f"Written Bronze table: {target_table}")
```

```python
# Validate Bronze weather table

display(spark.table(target_table))
print("Row count:", spark.table(target_table).count())
```
