# 02_governance_notes_and_permissions_limitations

## Purpose

Document Unity Catalog governance design and Free Edition permission limitations.

This notebook records the intended governance model and explains any practical limitations encountered in the Databricks Free Edition environment.

## Expected notes

- catalog and schema organization
- table and view placement
- volume usage for landing and checkpoints
- Catalog Explorer inspection
- permission limitations if `GRANT` or `SHOW GRANTS` are blocked

## Main idea

Even when full permission management is limited, the project should still show clear governance awareness and explain how the lakehouse is organized.

```python
print("Unity Catalog governance model")

print("Catalog: vattenfall_dev")
print("Schema raw: Bronze layer and governed landing/checkpoint volumes")
print("Schema refined: Silver cleaned and reusable datasets")
print("Schema analytics: Gold business-facing tables and views")
print("Volumes: landing and checkpoints are stored under the raw schema")
```

```python
catalog = "vattenfall_dev"

print("Schemas in catalog:")
display(spark.sql(f"SHOW SCHEMAS IN {catalog}"))
```

```python
analytics_schema = "analytics"

print("Analytics schema objects:")
display(spark.sql(f"SHOW TABLES IN {catalog}.{analytics_schema}"))
```

```python
gold_objects = [
    "gold_daily_market_summary",
    "gold_weather_impact_summary",
    "gold_grid_incident_summary",
    "gold_regional_operations_dashboard",
    "vw_regional_operations_dashboard",
    "vw_daily_incident_trends",
    "vw_market_weather_overview",
]

for object_name in gold_objects:
    full_name = f"{catalog}.{analytics_schema}.{object_name}"
    print(f"\nInspecting: {full_name}")
    display(spark.sql(f"DESCRIBE TABLE {full_name}"))
```

```python
print("Permissions and Free Edition note")

print("This project uses Unity Catalog to organize outputs into raw, refined, and analytics schemas.")
print("In a production environment, permissions would normally be managed with GRANT statements.")
print("In Databricks Free Edition, some permission commands may be limited or blocked.")
print("For this capstone, governance is demonstrated through catalog/schema/table/view organization and Catalog Explorer inspection.")
```

```python
print("Presentation explanation")

print("The project uses Unity Catalog as the governed delivery frame.")
print("Bronze tables live in raw because they preserve source arrivals.")
print("Silver tables live in refined because they are cleaned and reusable.")
print("Gold tables and views live in analytics because they are business-facing and dashboard-ready.")
print("This makes the lakehouse structure easier to inspect, explain, and govern.")
```
