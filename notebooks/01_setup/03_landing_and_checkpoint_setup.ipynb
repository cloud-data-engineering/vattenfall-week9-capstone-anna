```python
# Day 1 landing and checkpoint path setup for the Vattenfall capstone project

catalog = "vattenfall_dev"
schema = "raw"
landing_volume = "landing"
checkpoint_volume = "checkpoints"

landing_base_path = f"/Volumes/{catalog}/{schema}/{landing_volume}"
checkpoint_base_path = f"/Volumes/{catalog}/{schema}/{checkpoint_volume}"

landing_paths = {
    "market_prices": f"{landing_base_path}/market_prices",
    "weather": f"{landing_base_path}/weather",
    "grid_events": f"{landing_base_path}/grid_events",
    "reference": f"{landing_base_path}/reference",
}

checkpoint_paths = {
    "market_prices": f"{checkpoint_base_path}/market_prices_checkpoint",
    "weather": f"{checkpoint_base_path}/weather_checkpoint",
    "grid_events": f"{checkpoint_base_path}/grid_events_checkpoint",
    "reference": f"{checkpoint_base_path}/reference_checkpoint",
}

print("Landing base path:", landing_base_path)
print("Checkpoint base path:", checkpoint_base_path)

print("\nLanding paths:")
for domain, path in landing_paths.items():
    print(domain, "->", path)

print("\nCheckpoint paths:")
for domain, path in checkpoint_paths.items():
    print(domain, "->", path)
```

```python
# Create landing and checkpoint directories in Unity Catalog volumes

for path in landing_paths.values():
    dbutils.fs.mkdirs(path)
    print("Created landing path:", path)

for path in checkpoint_paths.values():
    dbutils.fs.mkdirs(path)
    print("Created checkpoint path:", path)
```

```python
# Verify landing and checkpoint directories

print("Landing directories:")
display(dbutils.fs.ls(landing_base_path))

print("Checkpoint directories:")
display(dbutils.fs.ls(checkpoint_base_path))
```

```python
# Copy sample CSV files from the repository to Unity Catalog landing directories

repo_path = "/Workspace/Users/dirella@startsteps.org/vattenfall-week9-capstone-anna"
sample_data_path = f"{repo_path}/sample_data"

source_to_landing = {
    "market_prices": landing_paths["market_prices"],
    "weather": landing_paths["weather"],
    "grid_events": landing_paths["grid_events"],
    "reference": landing_paths["reference"],
}

for domain, target_path in source_to_landing.items():
    source_path = f"{sample_data_path}/{domain}"

    print(f"\nCopying files for domain: {domain}")
    print("Source:", source_path)
    print("Target:", target_path)

    files = dbutils.fs.ls(source_path)

    for file in files:
        if file.name.endswith(".csv"):
            target_file = f"{target_path}/{file.name}"
            dbutils.fs.cp(file.path, target_file, True)
            print("Copied:", file.name)
```

```python
# Verify copied landing files

for domain, path in landing_paths.items():
    print(f"\nLanding files for {domain}:")
    display(dbutils.fs.ls(path))
```
